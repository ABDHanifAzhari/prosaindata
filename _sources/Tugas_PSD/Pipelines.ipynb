## **Tugas Pipeline**

Membuat alur pipelines :


1. Melakukan login ke https://cloud.orchest.io/
2. Membuat file project terlebih dahulu pada menu home  
3. Masuk kedalam file *project name* yang telah dibuat sebelumnya
4. Membuat model pipelines sesuai dengan yang ditugaskan dan jangan lupa untuk menghubungkan antar step
5. Melakukan akses ke https://www.cloudclusters.io/ untuk melakukan online data yang nantinya akan digunakan untuk melokalkan data yang akan diakses nantinya. **free 1 week**
6. Mengambil data (sepalwidth) yang diinginkan pada file csv di postgresql elephant.
7. Mengambil data (sepallength) yang diinginkan pada file csv dengan posgresql
8. Mengambil data (petallength) yang diinginkan pada file csv dengan mysql
9. Mengambil data (petalwidth dan class) yang diinginkan pada file csv dengan sql server (SSMS)
10. Menggabungkan data yang telah dionlinekan pada masing-masing step sebelumnya.
11. Melakukan penghitungan tingkat keakuratan dengan cara menggunakan metode seperti Decision Tree dan SVM

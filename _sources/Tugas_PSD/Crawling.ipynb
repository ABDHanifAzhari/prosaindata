# Crawling PTA

```python
import requests as req
from bs4 import BeautifulSoup as bs
import csv

hades = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'}
header = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'}
```

```python
#create an object HTTPresponse
def scrap_pta(hal):
  global hades
  a=0
  data=[]
  for page in range(1,hal):
        url = f"https://pta.trunojoyo.ac.id/c_search/byprod/10/{page}"
        request = req.get(url,hades).text
        soup = bs(request, 'html.parser')
        prodi = soup.find_all('div',{'id':'begin'})
        for pro in prodi:
          prod = pro.find('h2').text
          jur = prod[-18:]
          utama = soup.find('ul', 'items list_style')
          perid = utama.find_all('li', {'data-id':'id-1'})
          for x in perid:
            link = x.find('a','gray button')['href']
            request2 = req.get(link, hades).text
            soup2 = bs(request2, 'html.parser')
            abst= soup2.find('p',{'align':'justify'}).text.replace('\r','').replace('\n','')
            NPM = x.find('a','gray button')['href'][-12:]
            headline = x.find('a', 'title').text.replace('\r','').replace('\n','')
            print(f'done[{a}] > {headline[0:10]}')
            a+=1
            data.append([NPM,headline,abst,jur])
            kepala = ['NPM','Judul','Abstrak']
            writer = csv.writer(open('/content/drive/MyDrive/prosaindata/tugas/Tugas_PSD/crawling_pta.csv', 'w', newline=''))
            writer.writerow(kepala)
            for datas in data :
              writer.writerow(datas)
```

```python
scrap_pta(172)
```

# Crawling Berita

```python
import requests as req
from bs4 import BeautifulSoup as bs
import csv

header = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'}
```

```python
def scrap_detik(hal):
    global header
    a = 1
    data=[]
    for page in range(1,hal):
        url = f'https://www.detik.com/search/searchall?query=prabowo&siteid=2&sortby=time&page={page}'
        ge = req.get(url,header).text
        sop = bs(ge,'html.parser')
        list_berita = sop.find('div',class_='list media_rows list-berita')
        list_artikel = list_berita.find_all('article')
        for x in list_artikel:
            link = x.find('a')['href']
            judul = x.find('a').find('h2').text
            ge_ = req.get(link,header).text
            sop_ = bs(ge_,'html.parser')
            content = sop_.find_all('div',class_='detail__body-text itp_bodycontent')
            for x in content:
                x = x.find_all('p')
                y  = [y.text for y in x ]
                isi = ''.join(y).replace('\n', '').replace('ADVERTISEMENT','').replace('SCROLL TO RESUME CONTENT','')
                print(f'done[{a}] > {judul[0:10]}')
                a += 1
                data.append([judul,isi])
                kepala = ['Judul', 'Isi']
                writer = csv.writer(open('/content/drive/MyDrive/prosaindata/tugas/Tugas_PSD/crawling_berita.csv', 'w', newline=''))
                writer.writerow(kepala)
                for datas in data :
                  writer.writerow(datas)
```

```python
scrap_detik(8)
```

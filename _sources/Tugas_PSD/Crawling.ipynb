```python
import twint
import nest_asyncio
import pandas as pd
nest_asyncio.apply()

c = twint.Config()
c.Search = ['#AniesBaswedan']
c.Limit = 500
c.Pandas = True
# c.Store_csv = True
# c.Output = "./content/drive/MyDrive/prosaindata/tugas/Tugas_PSD/dataAnis.csv"
# twint.storage.panda.save
twint.run.Search(c)
```

```python
def column_names():
  return twint.output.panda.Tweets_df.columns

def twint_to_pd(columns):
  return twint.output.panda.Tweets_df[columns]

print(column_names())
```

```python
data = twint_to_pd(['created_at', 'username', 'tweet'])
print(data)

data.to_csv("/content/drive/MyDrive/prosaindata/tugas/Tugas_PSD/dataAnis.csv", index=False)
```
